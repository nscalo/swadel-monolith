### Benchmarking a model

Major benchmarking factors are: 

Throughput: which is the number of inferences per second of model execution, 
Latency: a threshold value in milliseconds time taken to deliver the model, useful for understanding bottlenecks, and 
Average Duration: The total duration per number of iterations

Average duration takes iunto consideration the cache latency, henmce throughput and latency are the best measures for evaluating the model.

Here, in this code, for small models such as human-pose-estimation-0001/FP16 nstreams value must be equated to zero, 

The code to benchmark a given model has been provided below:

```python
%run /opt/intel/openvino/deployment_tools/tools/benchmark_tool/benchmark_app.py \
-m /home/workspace/intel/human-pose-estimation-0001/FP16/human-pose-estimation-0001.xml \
-d CPU -api async -niter 10 -nireq 10 -progress -pc \
-i /home/workspace/images/sitting-on-car.jpg -nstreams
```
